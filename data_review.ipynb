```python
shibors=pd.read_csv("mfd_bank_shibor.csv",parse_dates=['mfd_date'],date_parser=lambda date: pd.datetime.strptime(date,"%Y%m%d"))
shibors=shibors.set_index("mfd_date")
shibors
```

```python
from matplotlib.dates import DayLocator,DateFormatter
fig=plt.gcf()
fig.set_size_inches(20,20)
plt.plot(shibors.Interest_O_N)
plt.gca().xaxis.set_major_locator(DayLocator(interval=15))
plt.gca().xaxis.set_major_formatter(DateFormatter("%y%m%d"))
plt.gcf().autofmt_xdate()
plt.show()
```

```python
fig=plt.gcf()
fig.set_size_inches(20,10)
plt.plot(shibors.Interest_O_N,color='r',label="隔夜利率")
plt.plot(shibors.Interest_1_W,color='g',label='1周利率')
plt.plot(shibors.Interest_1_M,color='b',label='1个月利率')
plt.plot(shibors.Interest_1_Y,color='y',label='1年利率')
plt.xlabel(u"日期")
plt.ylabel(u"利率")
plt.legend()
```

```python
interests=pd.read_csv("mfd_day_share_interest.csv",parse_dates=['mfd_date'],date_parser=lambda date: pd.datetime.strptime(date,"%Y%m%d"))
interests=interests.set_index('mfd_date')
interests
```

```python
fig=plt.gcf()
fig.set_size_inches(20,10)
plt.plot(shibors.Interest_O_N,color='r',label="拆借-隔夜利率")
plt.plot(shibors.Interest_1_W,color='g',label="拆借-1周利率")
plt.plot(interests.mfd_daily_yield,color='b',label="余额宝-万份收益")
plt.plot(interests.mfd_7daily_yield,color='y',label="余额宝-七日年化收益")
plt.xlabel(u"日期")
plt.ylabel(u"利率")
plt.legend()
```

```python
balances=pd.read_csv("user_balance_table.csv",parse_dates=['report_date'],date_parser=lambda date: pd.datetime.strptime(date,"%Y%m%d"))
```

```python
balances=balances.fillna(0).groupby(by="report_date").sum()
```

```python
balances["total_purchase_amt"].plot(figsize=(16,7),label="购买量",color='r')
balances["total_redeem_amt"].plot(figsize=(16,7),label="赎回量",color='b')
plt.legend()
```

```python
balances["total_purchase_amt"][274:304].plot(figsize=(16,7))
```

```python
balances["total_purchase_amt"].rolling(7).mean().plot(figsize=(16,7))
```

```python
balances["total_purchase_amt"].ewm(10).mean().plot()
```

```python
balances["total_purchase_amt"][-60:].plot(figsize=(16,7))
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
x=balances["total_purchase_amt"]
nx=(x-x.min())/(x.max()-x.min())
plot_acf(nx[-40:]).show()
```

```python
plot_pacf(x).show()
```
